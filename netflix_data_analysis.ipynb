## Insights from the data

##### Content Distribution: Movies vs. TV Shows on Netflix

```python
type_counts = df['type'].value_counts()
print(type_counts)
```

```python
# DATA VISUALIZATION
type_counts.plot(kind='pie', colors=['deeppink', 'purple'], autopct='%.2f%%', startangle=90)
plt.title('Content Distribution')
plt.ylabel('')
plt.show()
```

##### Common Ratings: Most popular ratings for content

```python
# Using mode() to directly get the required output
most_common_movie_rating = df[df.type == 'Movie']['rating'].mode().values[0]
print(f"Most common rating for Movies: {most_common_movie_rating}")

# Using value_counts() to see the breakdown of all ratings along with their frequencies
most_common_tv_show_rating = df[df['type']=='TV Show']['rating'].value_counts()
print(f"Most common rating for TV Shows: {most_common_tv_show_rating.idxmax()}")
```

##### Year-Specific Releases: Focus on movies from a particular year (e.g., 2020)

```python
movies_2020 = df[(df['type']=='Movie') & (df['release_year']==2020)]
print(movies_2020[['title','release_year']].reset_index(drop=True))
```

##### Top Content-Producing Countries: Leading countries by Netflix catalog size

```python
df['country'].head(10)
# By inspecting the data visually, we can see that there are NaN values present in the data & also there are multiple countries in a single entry
# creating a copy of data 
df_copy = df.copy()
df_copy
# Cleaning & splitting the data
# 1. Replacing NaN values with "Unknown"
df_copy['country'] = df_copy['country'].fillna('Unknown')
df_copy['country'] = df_copy['country'].replace('', 'Unknown')
# 2. Split the countries (if multiple countries are listed in one entry)
df_copy['country'] = df_copy['country'].str.split(',')    # split by commas
# 3. Clean the normalize the country names
import re
df_copy['country'] = df_copy['country'].apply(lambda x: [re.sub(r'[^a-zA-Z\s]', '', c.strip()).title() for c in x])
# 4. Flatten the list of countries
country_list = df_copy['country'].explode()
# 5. Count the occurrences of each country
top_10_countries = country_list[country_list.values != 'Unknown'].value_counts().head(10)
print(top_10_countries)
```

```python
# DATA VISUALIZATION
top_10_countries.plot(kind='barh', color='lightseagreen')
plt.gca().invert_yaxis()
plt.title('Top 10 Countries with the Most Content on Netflix')
plt.xlabel('Number of Titles')
plt.ylabel('Country')
plt.show()
```

##### Longest Movie: The movie with the maximum runtime

```python

# 1. Filtering Movie rows
df_movies = df[df['type']=='Movie']
# 2. Handling null values present in the duration column
df_movies = df_movies.dropna(subset='duration')
df_movies.duration.isnull().any()
# df_movies['duration'] = df_movies['duration'].fillna('0 min')   -- if we want to replace null values
df_movies
# We can see the duration column contains text data so we need to extract the numeric part from the data & convert it to int 
# 3. Extracting the numeric part 
df_movies['duration_min'] = df_movies['duration'].str.split(' ').str[0].astype(int)  # using str.split()
# df[movies_duration] = df['duration'].str.extract(r'(\d+)').astype(int)                # using str.extract()
# 4. Filter out the longest movie
longest_movie = df_movies.loc[df_movies['duration'].idxmax()]
longest_movie[['title', 'description', 'duration_min']]
```

##### Genre Popularity: Genre-wise content distribution

```python
df.head()
df_copy = df.copy()
# By inspecting the data visually, we can observe that the 'listed_in' column contains multiple genres in a single entry separated by comma
df_copy['listed_in'].isna().any()   # checking for null values
# 1. Split the genre's (if multiple genre's are listed in one entry)
df_copy['listed_in'] = df_copy['listed_in'].str.split(',')  # split by comma
# 2. Clean & normalize the genre names
df_copy['listed_in'] = df_copy['listed_in'].apply(lambda x: [re.sub(r'[^a-zA-Z\s]', '', c.strip()).title() for c in x])
# 3. Flatten the list of genre's
genre_list = df_copy['listed_in'].explode()
# 4. Count the content items in each genre
genre_counts = genre_list.value_counts().reset_index()
genre_counts.columns = ['Genre', 'Count']
genre_counts
```

```python
# DATA VISUALIZATION
plt.figure(figsize=(12, 10))
sns.barplot(genre_counts, x='Count', y='Genre', hue='Genre', palette='viridis')
plt.title('Content Distribution of Genres')
plt.xlabel('Number of Content Items')
plt.ylabel('Genres')
plt.show()
```

##### India's Content Growth: Top years for content releases in India

```python
# Filter data of India
df_india = df[df['country'].str.contains('India', na=False)]
# Group by release_year & find the total releases per year
content_per_year = df_india.groupby('release_year').size().reset_index(name='content_count').sort_values(by='content_count', ascending=False)
top_5 = content_per_year.head(5)
print(top_5)
```

```python
# DATA VISUALIZATION
plt.figure(figsize=(10, 6))
sns.barplot(data=top_5, x='release_year', y='content_count', palette='plasma', order=top_5['release_year'])
plt.title('Top 5 Years with the Highest Content Releases in India', fontsize=16)
plt.xlabel('Release Year', fontsize=14)
plt.ylabel('Content Count', fontsize=14)
plt.show()
```

##### Top Actors in Indian Movies: Most frequent actors in Indian movies

```python
# 1. Filtering data for Movies produced in India
movies_india = df[(df['type']=='Movie') & (df['country'].str.contains('India', na=False))].copy()
# As the cast col contains multiple actors clubbed in a single entry, we need to clean this data first
# 2. Split the cast col into individual actors & then flatten the list
movies_india['cast'] = movies_india['cast'].dropna().str.split(',').apply(lambda x: [actor.strip() for actor in x])
actors_list = movies_india['cast'].explode()
# 3. Count the total movies by each actor
top_10_actors = actors_list.value_counts().reset_index().head(10)
top_10_actors.columns = ['actor', 'movie_count']
print(top_10_actors)
```

```python
# DATA VISUALIZATION
plt.figure(figsize=(10, 6))
sns.barplot(data=top_10_actors, x='movie_count', y='actor', color='teal')
plt.title('Top 10 Actors with highest number of movies produced in India')
plt.xlabel('Total Movies')
plt.ylabel('Actor')
plt.show()
```

##### Thematic Categorization: Classifying content as 'Good' or 'Bad' based on keywords 
(e.g. 'Bad' if keywords like "kill" & "violence" present in the description field & all remaining as 'Good')

```python
df['category'] = np.where(df['description'].str.contains('kill|violence', case=False, na=False), 'Bad', 'Good')
category_count = df['category'].value_counts().reset_index()
category_count.columns = ['Category', 'Count']
print(category_count)
```

```python
# DATA VISUALIZATION
plt.figure(figsize=(10, 4))
sns.barplot(x='Count', y='Category', data=category_count, hue='Category', palette=['lightcoral', 'brown'], orient='h')
plt.title('Content Categorization Based on Keywords', fontsize=16)
plt.xlabel('Count', fontsize=12)
plt.ylabel('Category', fontsize=12)
plt.show()
```

##### Content Production Growth: Netflix’s scaling over the years

```python
# Convert the 'date_added' column to datetime, handle null values and extract the year
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df_cleaned = df.dropna(subset='date_added')
df_cleaned['year_added'] = df_cleaned['date_added'].dt.year
# Separate the data based on 'type'
movies = df_cleaned[df_cleaned['type']=='Movie'].groupby('year_added').size().reset_index(name='Movies')
tv_shows = df_cleaned[df_cleaned['type']=='TV Show'].groupby('year_added').size().reset_index(name='TV Shows')
# Merge both the data for better alignment (not to skip any missing year in one type)
combined = pd.merge(movies, tv_shows, how='outer', on='year_added').fillna(0)

# DATA VISUALIZATION
plt.figure(figsize=(15, 10))
sns.lineplot(combined, x='year_added', y='Movies', marker='o', label='Movies', color='maroon')
sns.lineplot(combined, x='year_added', y='TV Shows', marker='o', label='TV Shows', color='purple')
plt.title('Netflix Content Production over Time: Movies vs TV Shows')
plt.xlabel('Year')
plt.ylabel('Total Content')
plt.grid(axis='y')
plt.show()
```
